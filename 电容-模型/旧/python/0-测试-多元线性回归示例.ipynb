```python
import pandas as pd
import numpy as np

data = pd.read_excel('all_label.xlsx')

features = data.iloc[:, :8].values
labels = data.iloc[:, 8].values
```

```python
features
```

```python
labels
```

```python
# 添加偏置项
features = np.c_[np.ones(features.shape[0]), features]

# 计算参数 theta
theta = np.linalg.inv(features.T @ features) @ features.T @ labels

theta
```

```python
# 预测
data2 = pd.read_excel('test_label.xlsx')

test_features = data2.iloc[:, :8].values
test_features = np.c_[np.ones(test_features.shape[0]), test_features]
predictions = np.dot(test_features, theta)

# 输出预测结果
predictions
```
