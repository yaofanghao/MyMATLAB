```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('loss.xlsx', usecols=[0, 1, 2, 3, 4])

start_num = 10
end_num = 2000

col1 = data.iloc[start_num:end_num, 0]
col2 = data.iloc[start_num:end_num, 2]
col3 = data.iloc[start_num:end_num, 1]
col4 = data.iloc[start_num:end_num, 3]
col4 = data.iloc[start_num:end_num, 4]

fig, ax = plt.subplots()

ax.plot(col1, label='bp-100')
ax.plot(col2, label='bp-100-5layers')
ax.plot(col3, label='bp-200')
ax.plot(col4, label='resnet-128-128')
ax.plot(col4, label='resnet-128-256')

ax.legend()
ax.set_title('model loss comparison')
ax.set_xlabel('epoch')
ax.set_ylabel('loss')

plt.show()
```
